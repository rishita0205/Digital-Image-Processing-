```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("/content/me.jpg")
img1=cv2.resize(img1,(224,224))
H,W=img1.shape[:2]
h_img=H/2;
w_img=W/2;
```

```python
T=np.float32([[1,0,h_img], [0,1,w_img]])
trans_img=cv2.warpAffine(img1,T,(H,W))
```

```python
fig=plt.figure(figsize=(10,7))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1,2,2)
plt.imshow(trans_img)
plt.axis("off")
plt.title("translated image")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from cv2 import ROTATE_90_CLOCKWISE

img2=cv2.imread("/content/me.jpg")
img2=cv2.resize(img2,(224,224))
H,W=img2.shape[:2]
h_img2,w_img2 =H/2,W/2
rotated_img = cv2.warpAffine(img2, cv2.getRotationMatrix2D((w_img2, h_img2), 45, 0.5), (W, H), flags=ROTATE_90_CLOCKWISE)
```

```python
fig=plt.figure(figsize=(10,7))
fig.add_subplot(1,2,1)
plt.imshow(img2)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1,2,2)
plt.imshow(rotated_img)
plt.axis("off")
plt.title("rotated image")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img3 = cv2.imread("/content/me.jpg")
img3 = cv2.resize(img3,(244,244))
H,W=img3.shape[:2]
T=np.float32([[1,0,0],[1,1,0],[0,0,1]])

x_shear = cv2.warpPerspective(img3 , T , (int(H*1.5) , int(W*1.5)))
```

```python
fig = plt.figure(figsize = ( 10, 7))
fig.add_subplot(1 , 2 , 1)
plt.imshow(img3)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1 , 2 , 2)
plt.imshow(x_shear)
plt.axis("off")
plt.title("sheared image")
```
