```python
import matplotlib.pyplot as plt
import numpy as np # Import NumPy # Read the image
img1 = cv2.imread("/content/rishita.png")
# Convert the image to grayscale
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) # Display the grayscale image
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show() # Hide axis
```

```python
img = img2

lst = []

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    lst.append(np.binary_repr(img[i][j],width=8))

img_bit_8 = (np.array([int(i[0]) for i in lst], dtype = np.uint8) *
128).reshape(img.shape[0],img.shape[1])
img_bit_7 = (np.array([int(i[1]) for i in lst], dtype = np.uint8) *
64).reshape(img.shape[0],img.shape[1])
img_bit_6 = (np.array([int(i[2]) for i in lst], dtype = np.uint8) *
32).reshape(img.shape[0],img.shape[1])
img_bit_5 = (np.array([int(i[3]) for i in lst], dtype = np.uint8) *
16).reshape(img.shape[0],img.shape[1])
img_bit_4 = (np.array([int(i[4]) for i in lst], dtype = np.uint8) *
8).reshape(img.shape[0],img.shape[1])
img_bit_3 = (np.array([int(i[5]) for i in lst], dtype = np.uint8) *
4).reshape(img.shape[0],img.shape[1])
img_bit_2 = (np.array([int(i[6]) for i in lst], dtype = np.uint8) *
2).reshape(img.shape[0],img.shape[1])
img_bit_1 = (np.array([int(i[7]) for i in lst], dtype = np.uint8) *
1).reshape(img.shape[0],img.shape[1])
fig= plt.figure(figsize=(10,7))
fig.add_subplot(2,4,1)
plt.imshow(img_bit_8, cmap='gray')
plt.axis("off")
plt.title("8_bit_image")

fig.add_subplot(2,4,2)
plt.imshow(img_bit_7, cmap='gray')
plt.axis("off")
plt.title("7_bit_image")
fig.add_subplot(2,4,3)
plt.imshow(img_bit_6, cmap='gray')
plt.axis("off")
plt.title("6_bit_image")
fig.add_subplot(2,4,4)
plt.imshow(img_bit_5, cmap='gray')
plt.axis("off")
plt.title("5_bit_image")
fig.add_subplot(2,4,5)
plt.imshow(img_bit_4, cmap='gray')
plt.axis("off")
plt.title("4_bit_image")
fig.add_subplot(2,4,6)
plt.imshow(img_bit_3, cmap='gray')
plt.axis("off")
plt.title("3_bit_image")
fig.add_subplot(2,4,7)
plt.imshow(img_bit_2, cmap='gray')
plt.axis("off")
plt.title("2_bit_image")
fig.add_subplot(2,4,8)
plt.imshow(img_bit_1, cmap='gray')
plt.axis("off")
plt.title("1_bit_image")
plt.show()
```
