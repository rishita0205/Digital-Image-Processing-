```python
import cv2                                                                               # Importing the OpenCV library
img1 = cv2.imread("/content/me.jpg")                       # Reading an image file named "me.jpeg"
from google.colab.patches import cv2_imshow              # Importing a function to display images in Colab
cv2_imshow(img1)                                                                # Displaying the image using the imported function
img1.shape                                                                             # Retrieving and printing the shape of the image

```

```python
H=255                                                                          # Set height to 255
W=255                                                                        # Set width to 255
img3=cv2.resize(img1,(200,200))                          # Resize img1 to (200,200) and assign to img3
img3.shape                                                                # Display the shape of img3
img3.size                                                                   # Display the size of img3
```

```python
import matplotlib.pyplot as plt                               # Importing the pyplot module from matplotlib
plt.imshow(img1)                                                      # Displaying the image using the imshow function
plt.axis("off")                                                              # Turning off the axis for a cleaner display
```

```python
# resize the image
img4=cv2.resize(img1,(100,100))                        # Resizing img1 to 100x100
img5=cv2.resize(img1,(50,50))                             # Resizing img1 to 50x50
fig = plt.figure(figsize=(10,7))                                # Create a figure with a size of 10x7 inches
fig.add_subplot(1,3,1)                                           # Add a subplot (1 row, 3 columns), and set the   curr
plt.imshow(img1)                                                   # Display the original image in the first subplot
plt.axis("off")                                                           # Turn off axis labels and ticks
plt.title("Original image ")                                     # Set the title for the first subplot

fig.add_subplot(1,3,2)                                            # Move to the second subplot
plt.imshow(img4)                                                    # Display the image with size 100x100 in the second subplot
plt.axis("off")                                                           # Turn off axis labels and ticks
plt.title("Size=100x100 ")                                      # Set the title for the second subplot

fig.add_subplot(1,3,3)                                           # Move to the third subplot
plt.imshow(img5)                                                   # Display the image with size 50x50 in the third subplot
plt.axis("off")                                                           # Turn off axis labels and ticks
plt.title("Size=50x50 ")                                          # Set the title for the third subplot
```

```python
img6 = img1[:,:,0]                                     # Extracting the first channel (red) from the image
img7 = img1[:,:,1]                                     # Extracting the second channel (green) from the image
img8 = img1[:,:,2]                                     # Extracting the third channel (blue) from the image

fig=plt.figure(figsize=(10,7))                     # Create a figure with a size of 10x7 inches
fig.add_subplot(1,3,1)                               # Add a subplot to the figure with 1 row, 3 columns, and index 1
plt.axis("off")                                               # Turn off axis ticks and labels
plt.imshow(img6)                                       # Display the image 'img6'
plt.title("R_gray")                                       # Set the title of the subplot to "R_gray"

fig.add_subplot(1,3,2)                                # Add a subplot to a 1x3 grid at position 2
plt.axis("off")                                                # Turn off axis labels and ticks
plt.imshow(img7)                                         # Display the image img7
plt.title("G_gray")                                        # Set the title of the subplot to "G_gray"

fig.add_subplot(1,3,3)                                # Add a subplot to the figure with 1 row, 3 columns, and index 3
plt.axis("off")                                                # Turn off axis labels and ticks
plt.imshow(img8)                                        # Display the image img8
plt.title("B_gray")                                        # Set the title of the subplot to "B_gray"
```

```python
b, g, r = cv2.split(img1)                                                    # Split the image into its RGB channels
# Perform channel subtractions
rg_subtraction = r - g
br_subtraction = b - r
bg_subtraction = b - g

fig = plt.figure(figsize=(10, 7))                                         # Display the original and subtracted images
# Subtraction: R - G
plt.subplot(1, 3, 1)                                                            # Display the R - G subtraction
plt.imshow(rg_subtraction, cmap='coolwarm')
plt.axis("off")
plt.title("R - G")

# Subtraction: B - R
plt.subplot(1, 3, 2)                                                            # Display the B - R subtraction
plt.imshow(br_subtraction, cmap='coolwarm')
plt.axis("off")
plt.title("B - R")

# Subtraction: B - G
plt.subplot(1, 3, 3)                                                            # Display the B - G subtraction
plt.imshow(bg_subtraction, cmap='coolwarm')
plt.axis("off")
plt.title("B - G")
plt.show()
```

```python
# Extract the red and green channels from the original image
img6 = img1[:, :, 0]  # Red channel
img7 = img1[:, :, 1]  # Green channel
# Resize the red and green channel images for better visualization
img6 = cv2.resize(img6, (200, 200))
img7 = cv2.resize(img7, (200, 200))
# Perform basic arithmetic operations on the red and green channels
img_add = cv2.add(img6, img7)  # Addition
img_sub = cv2.subtract(img6, img7)  # Subtraction
img_mul = cv2.multiply(img6, img7)  # Multiplication
img_div = cv2.divide(img6, img7)  # Division
fig, axs = plt.subplots(2, 2)                                     # Create a 2x2 subplot for visualization

# Display the result after addition
axs[0, 0].imshow(img_add)
axs[0, 0].set_title("Addition")
axs[0, 0].axis("off")

# Display the result after subtraction
axs[0, 1].imshow(img_sub)
axs[0, 1].set_title("Subtraction")
axs[0, 1].axis("off")

# Display the result after multiplication
axs[1, 0].imshow(img_mul)
axs[1, 0].set_title("Multiplication")
axs[1, 0].axis("off")

# Display the result after division
axs[1, 1].imshow(img_div)
axs[1, 1].set_title("Division")
axs[1, 1].axis("off")
plt.subplots_adjust(wspace=0.5, hspace=0.5)              # Adjust spacing between subplots for better layout
plt.show()
```
