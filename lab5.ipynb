```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_array = np.array([[1, 2, 3, 4, 5],
                        [6, 7, 0, 1, 2],
                        [3, 4, 5, 6, 7],
                        [0, 1, 2, 3, 4],
                        [5, 6, 7, 0, 1]])
```

```python
def zero_padding(arr,padding_size):
  return np.pad(arr,pad_width=padding_size, mode='constant',constant_values=(0))
```

```python
zero_padded_arr = zero_padding(image_array,2)
zero_padded_arr
```

```python
plt.imshow(zero_padded_arr)
```

```python
img1 = cv2.imread("/content/me.jpg")
img1 = cv2.resize(img1, (224, 224))
```

```python
zero_padded_img = cv2.copyMakeBorder(img1, 20, 20, 20, 20, cv2.BORDER_CONSTANT, None, value =0)

fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1,2,2)
plt.imshow(zero_padded_img)
plt.axis("off")
plt.title("zero_padded_img")
```

```python
image_array = np.array([[1, 2, 3, 4, 5],
                        [6, 7, 0, 1, 2],
                        [3, 4, 5, 6, 7],
                        [0, 1, 2, 3, 4],
                        [5, 6, 7, 0, 1]])

def mirror_padding(arr, padding_size):
  return np.pad(arr,pad_width=padding_size, mode='reflect')
```

```python
mirror_padded_arr = mirror_padding(image_array, 2)
print("\nMirror Padded Array:")
print(mirror_padded_arr)
```

```python
img1 = cv2.imread("/content/me.jpg")
img1 = cv2.resize(img1, (224, 224))
```

```python
Mirror_pad_img = cv2.copyMakeBorder(img1, 20, 20, 20, 20, cv2.BORDER_REFLECT)

fig = plt.figure(figsize=(6,4))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1,2,2)
plt.imshow(Mirror_pad_img)
plt.axis("off")
plt.title("Mirror_padded_img")
```

```python
image_array = np.array([[1, 2, 3, 4, 5],
                        [6, 7, 0, 1, 2],
                        [3, 4, 5, 6, 7],
                        [0, 1, 2, 3, 4],
                        [5, 6, 7, 0, 1]])

```

```python
def replicate_padding(arr, padding_size):
  return np.pad(arr, pad_width=padding_size, mode="edge")
```

```python
replicate_padded_arr = replicate_padding(image_array, 1)
```

```python
 print("\nMirror Padded Array:")
print(replicate_padded_arr)
```

```python
replicate_padding_img = cv2.copyMakeBorder(img1, 20, 20, 20, 20, cv2.BORDER_REPLICATE)

fig = plt.figure(figsize=(6,4))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.axis("off")
plt.title("original image")

fig.add_subplot(1,2,2)
plt.imshow(replicate_padding_img)
plt.axis("off")
plt.title("replicate_padded_img")
```
