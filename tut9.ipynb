```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 3x3 matrix
matrix = np.array([[167, 133, 133],
                   [144, 140, 135],
                   [159, 154, 148]],dtype=np.uint8)

plt.imshow(matrix, cmap="gray")
plt.show()

# Convert each element to 8-bit binary representation
binary_matrix = np.array([[int(np.binary_repr(x, 8)) for x in row] for row in matrix])
print(binary_matrix)


# Create 8 bit planes
bit_planes = []
for i in range(8):
    bit_planes.append(np.right_shift(binary_matrix, i) & 1)

# Print the bit planes
for i, bit_plane in enumerate(bit_planes):
    print(f"Bit Plane {i+1}")
    plt.imshow(matrix, cmap="gray")
    plt.show()
```
