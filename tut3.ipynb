```python
# Code and Methodology:
import cv2
import matplotlib.pyplot as plt
import numpy as np

img1=cv2.imread("/content/Boat.jpg")
from google.colab.patches import cv2_imshow
```

```python
from PIL import Image
img2= Image.open("/content/me.jpg")
img3 = img2.convert("P",palette=Image.ADAPTIVE,colors=32)
img4 = img2.convert("P",palette=Image.ADAPTIVE,colors=16)
img5 = img2.convert("P",palette=Image.ADAPTIVE,colors=8)
plt.imshow(img3)

fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.imshow(img3)
plt.axis("off")
plt.title("Color_depth=32")

fig.add_subplot(1,3,2)
plt.imshow(img4)
plt.axis("off")
plt.title("Color_depth=16")

fig.add_subplot(1,3,3)
plt.imshow(img5)
plt.axis("off")
plt.title("Color_depth=8")

```

```python
img6= cv2.imread("/content/me.jpg")
img6= cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
W, H = img6.shape[:2]
img7 = cv2.imread("/content/Boat.jpg")
img7 = cv2.resize(img7,(H, W))
img7= cv2.cvtColor(img7,cv2.COLOR_BGR2GRAY)

```

```python
img8= np.logical_and(img6, img7)
img9 = Image.fromarray(img8)
plt.axis("off")

img10= np.logical_or(img6, img7)
img11 = Image.fromarray(img10)
plt.axis("off")

img12= np.logical_xor(img6, img7)
img13 = Image.fromarray(img12)
plt.axis("off")

img8= np.logical_and(img6, img7)
img9 = Image.fromarray(img8)

img10= np.logical_or(img6, img7)
img11 = Image.fromarray(img10)

img12= np.logical_xor(img6, img7)
img13 = Image.fromarray(img12)

img14=np.logical_not(img6)
img14 = Image.fromarray(img14)

img15=np.logical_not(img7)
img15 = Image.fromarray(img15)


```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,4,1)
plt.imshow(img6)
plt.axis("off")
plt.title("Image 1")

fig.add_subplot(2,4,2)
plt.imshow(img7)
plt.axis("off")
plt.title("Image 2")

fig.add_subplot(2,4,3)
plt.imshow(img9)
plt.axis("off")
plt.title("Logical_And")

fig.add_subplot(2,4,4)
plt.imshow(img11)
plt.axis("off")
plt.title("Logical_OR")

fig.add_subplot(2,4,5)
plt.imshow(img13)
plt.axis("off")
plt.title("Logical_XOR")

fig.add_subplot(2,4,6)
plt.imshow(img14)
plt.axis("off")
plt.title("Logical_Not")

fig.add_subplot(2,4,7)
plt.imshow(img15)
plt.axis("off")
plt.title("Logical_Not")

```

```python
img16= cv2.imread("/content/me.jpg")
W, H = img16.shape[:2]
img17 = cv2.imread("/content/Boat.jpg")
img18 = cv2.resize(img17,(H, W))
img_1=cv2.bitwise_and(img16,img18)

img_2= cv2.bitwise_or(img16,img18)

img_3= cv2.bitwise_xor(img16,img18)

img_4= cv2.bitwise_not(img16)

img_5= cv2.bitwise_not(img18)

```

```python
fig=plt.figure(figsize=(10,7))
fig.add_subplot(2,4,1)
plt.imshow(img16)
plt.axis("off")
plt.title("Image-1")

fig.add_subplot(2,4,2)
plt.imshow(img18)
plt.axis("off")
plt.title("Image-2")

fig.add_subplot(2,4,3)
plt.imshow(img_1)
plt.axis("off")
plt.title("AND_operation")

fig.add_subplot(2,4,4)
plt.imshow(img_2)
plt.axis("off")
plt.title("OR_operation")

fig.add_subplot(2,4,5)
plt.imshow(img_3)
plt.axis("off")
plt.title("XOR_operation")

fig.add_subplot(2,4,6)
plt.imshow(img_4)
plt.axis("off")
plt.title("NOT_operation")

fig.add_subplot(2,4,7)
plt.imshow(img_5)
plt.axis("off")
plt.title("NOT_operation")

```
